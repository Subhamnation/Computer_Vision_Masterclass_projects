## Detecting Faces with HOG and Dlib

```python
from google.colab.patches import cv2_imshow
image = cv2.imread('/content/drive/MyDrive/Computer Vision Masterclass/Images/people2.jpg')
cv2_imshow(image)
```

```python
face_detector_hog = dlib.get_frontal_face_detector()
```

```python
detections = face_detector_hog(image, 1)
```

```python
detections
```

```python
len(detections)
```

```python
for face in detections:
  print(face)
```

```python
for face in detections:
  #print(face)
  print(face.left())
  print(face.top())
  print(face.right())
  print(face.bottom())
```

```python
l,t,r,b = face.left(),face.top(),face.right(),face.bottom()
```

```python
for face in detections:
  l,t,r,b = face.left(),face.top(),face.right(),face.bottom()
  cv2.rectangle(image, (l,t), (r,b), (255,0, 255), 2 )
cv2_imshow(image)
```

```python
detections = face_detector_hog(image, 4)
```

```python
for face in detections:
  l,t,r,b = face.left(),face.top(),face.right(),face.bottom()
  cv2.rectangle(image, (l,t), (r,b), (255,255, 0), 2 )
cv2_imshow(image)
```

```python
detections = face_detector_hog(image, 3)
```

```python
for face in detections:
  l,t,r,b = face.left(),face.top(),face.right(),face.bottom()
  cv2.rectangle(image, (l,t), (r,b), (0,255, 255), 2 )
cv2_imshow(image)
```
